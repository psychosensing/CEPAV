```python
def extract_time_from_file(filename):
    try:
        with open(filename, 'r') as file:
            # Skip the first row
            next(file)
            # Read the second row
            second_row = next(file).strip()
            # Split by "/"
            parts = second_row.split("/")
            # Return the time part after "/"
            return parts[1]
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
        return None
```
